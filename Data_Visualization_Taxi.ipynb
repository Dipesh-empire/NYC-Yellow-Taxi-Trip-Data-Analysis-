```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df =sns.load_dataset("taxis")
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['totaltime'] = df['dropoff']-df['pickup']
```

```python
df.head()
```

```python
# Bivariate analysis 
plt.scatter(df['distance'],df['total'],color='blue',marker ='x')
plt.title("Distance and price comparision")
plt.xlabel("Distance")
plt.ylabel("total")
plt.show()
```

```python
# Almost 36% people didn't get any tips 
(df['tip'] == 0.0).sum() / len(df) * 100
```

```python
# Univariate analysis 
plt.hist(df['color'],bins=5)
plt.title("Color of the taxis")
plt.show()
```

```python
# Univariate analysis 

plt.bar(df['payment'].value_counts().index,df['payment'].value_counts().values)
plt.title("Payment type")
plt.show()
```

```python
#I got this insight from the distance and price comparison and as you can  see  the distance O have total 
df[df['distance']==0.00]
```

```python
# now dropping those rows 
df = df[df['distance'] != 0.00]
```

```python
# Bivariate analysis 
plt.scatter(df['distance'],df['total'],color='blue',marker ='x')
plt.title("Distance and price comparision")
plt.xlabel("Distance")
plt.ylabel("total")
plt.show()
```

```python
pivot = df.groupby(['pickup_borough', 'payment'])['tip'].mean().unstack()
pivot.plot(kind='bar', figsize=(10,6))
plt.title('Average Tip by Borough and Payment Type')
plt.ylabel('Average Tip')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='distance', y='fare', hue='payment')
plt.title('Distance vs Fare colored by Payment Type')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['distance'], df['fare'], df['tip'], c='blue', alpha=0.5)
ax.set_xlabel('Distance')
ax.set_ylabel('Fare')
ax.set_zlabel('Tip')
plt.title('3D Scatter: Distance vs Fare vs Tip')
plt.show()
```
